```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=25)
```

```python
fig2 = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax20 = fig2.add_subplot(4,1,r-3)
    ax20 = recent_grads[cols[r]].plot(kind='hist', rot=40)
```

```python
import numpy as np

EachGenderMajorCat = recent_grads.pivot_table(index='Major_category', values=['Men','Women'],aggfunc=sum)
EachGenderMajorCat
```

```python
fig3, ax30 = plt.subplots()
ax30.boxplot(recent_grads["Median"])
ax30.set_xticklabels("Median", rotation=90)
ax30.set_ylim(0,100000)
plt.show()
```
